```python
import requests
import json 
from bs4 import BeautifulSoup as bs
from webdriver_manager.chrome import ChromeDriverManager
import splinter 
import selenium
import pandas as pd
```

```python
mars_dict={}
```

```python
executable_path = {'executable_path' : 'chromedriver.exe'}
browser = splinter.Browser('chrome', **executable_path, headless=False)
url = 'https://mars.nasa.gov/news/'
browser.visit(url)
html = browser.html
```

```python
soup = bs(html, 'html.parser')
```

```python
results=soup.find_all('div', class_="content_title")[1]
```

```python
news_title=results.find('a').text.strip()
```

```python
news_p = soup.find('div', class_='article_teaser_body').text.strip()
news_p
```

```python
mars_dict["news title"]=news_title
mars_dict["news paragraph"]=news_p
```

```python
## Mars Hemisphere
hemisphere_url="https://astrogeology.usgs.gov/search/results?q=hemisphere+enhanced&k1=target&v1=Mars"
```

```python
browser.visit(hemisphere_url)
html = browser.html
```

```python
soup = bs(html, 'html.parser')
```

```python
divs=soup.find("div", class_="collapsible")
# hemisphere_img_urls=[]
```

```python
hemisphere_img_urls=[]
items=soup.find_all('div', class_='item')
name=[]
img_url=[]

for item in items:
    dict_s={}
    browser.visit(hemisphere_url) 
    title=item.find('h3').get_text()
    browser.click_link_by_partial_text(title)
    html=browser.html
    soup=bs(html, "html.parser")
    downloads=soup.find("div", class_="downloads")    img=browser.url 
    img=downloads.find("a")["href"]

    
    dict_s["title"] = title
    dict_s["img_url"]=img
            
    hemisphere_img_urls.append(dict_s)

mars_dict["Hemisphere urls"]=hemisphere_img_urls
    
mars_dict
```
